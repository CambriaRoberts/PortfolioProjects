```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

Basic Graph

```python
x = [1,3,5,7,9]

y = [2,4,6,9,13]

#Sizing Graph
plt.figure(figsize=(5,3), dpi = 120)

#plt.plot(x,y, label = '2x', marker = '.', markersize = 10, linestyle ='--', color = 'green')
plt.plot(x,y, 'g^--', label ='2x')

#Second line

x2 = np.arange(1,7.5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'k', label = 'X^2')
plt.plot(x2[4:], x2[4:]**2, 'k--')

plt.title('Generic Graph')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
plt.show()

plt.savefig('mygraph.png', dpi= 200)
```

Bar Charts

```python
labels = ['A', 'B', 'C']
values = [3,5,9]

bars = plt.bar(labels, values)

patterns = ['O', '*', '/']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```
